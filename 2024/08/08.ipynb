```python
def get_locations(fpath):
    locations = {}
    with open(fpath) as f:
        for i, line in enumerate(f):
            line = line.strip()
            for j, c in enumerate(line):
                if c != ".":
                    locations[c] = locations.get(c, []) + [(i, j)]
    maxi = i
    maxj = j
    return locations, maxi, maxj


def get_antinodes(loc1, loc2, maxi, maxj):
    antinodes = set()
    d12 = (loc1[0] - loc2[0], loc1[1] - loc2[1])
    # loc1 + d12
    a1 = (loc1[0] + d12[0], loc1[1] + d12[1])
    if not (a1[0] < 0 or a1[1] < 0 or a1[0] > maxi or a1[1] > maxj):
        antinodes.add(a1)
    # loc2 - d12
    a2 = (loc2[0] - d12[0], loc2[1] - d12[1])
    if not (a2[0] < 0 or a2[1] < 0 or a2[0] > maxi or a2[1] > maxj):
        antinodes.add(a2)
    return antinodes
```

```python
type_to_locations, maxi, maxj = get_locations("input.txt")

antinodes = set()
for type, locations in type_to_locations.items():
    for i in range(len(locations)):
        for j in range(i + 1, len(locations)):
            loc1 = locations[i]
            loc2 = locations[j]
            antinodes_pair = get_antinodes(loc1, loc2, maxi, maxj)
            antinodes.update(antinodes_pair)

print(len(antinodes))
print(antinodes)
```

```python
def get_antinodes_with_resonances(loc1, loc2, maxi, maxj):
    d12 = (loc1[0] - loc2[0], loc1[1] - loc2[1])
    resonances = set()
    # loc1 + d12
    a1 = [loc1[0], loc1[1]]
    while not (a1[0] < 0 or a1[1] < 0 or a1[0] > maxi or a1[1] > maxj):
        # the antenna is a resonance
        resonances.add((a1[0], a1[1]))
        a1[0] += d12[0]
        a1[1] += d12[1]

    # loc2 - d12
    a2 = [loc2[0], loc2[1]]
    while not (a2[0] < 0 or a2[1] < 0 or a2[0] > maxi or a2[1] > maxj):
        # the antenna is a resonance
        resonances.add((a2[0], a2[1]))
        a2[0] -= d12[0]
        a2[1] -= d12[1]

    return resonances


type_to_locations, maxi, maxj = get_locations("input.txt")
type_to_locations, maxi, maxj

antinodes = set()
for type, locations in type_to_locations.items():
    for i in range(len(locations)):
        for j in range(i + 1, len(locations)):
            loc1 = locations[i]
            loc2 = locations[j]
            antinodes_pair = get_antinodes_with_resonances(loc1, loc2, maxi, maxj)
            antinodes.update(antinodes_pair)


print(len(antinodes))
print(antinodes)
```
